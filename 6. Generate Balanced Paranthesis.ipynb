```python
def generateParenthesis(openB, closeB, n, s=[]):
    #Base Case
    if(closeB==n):
        print(''.join(s))
        return
    
    else:
        if(openB>closeB):
            #You can definately put one closing bracket
            s.append(')')
            generateParenthesis(openB, closeB+1, n, s)
            #Backtrack
            s.pop()
        if(openB<n):
            s.append('(')
            generateParenthesis(openB+1, closeB, n, s)
            s.pop()
    return

n = int(input())
generateParenthesis(0, 0, n)
```
